## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_1=pd.read_csv('./data/hist_data.csv')
plt.hist(df_1['aデータ'],color="orange",alpha=0.3,bins=30,label="aデータ")
plt.hist(df_1['bデータ'],color="turquoise",alpha=0.3,bins=30,label="bデータ")
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_2=pd.read_csv("./data/scatter_data.csv")
cluster0=df_2['class']==0
cluster1=df_2['class']==1
cluster2=df_2['class']==2
plt.grid()
plt.scatter('length (cm)','width (cm)',data=df_2[cluster0], c="r", marker="o",label="class0")
plt.scatter('length (cm)','width (cm)',data=df_2[cluster1], c="b", marker="o",label="class1")
plt.scatter('length (cm)','width (cm)',data=df_2[cluster2], c="g", marker="o",label="class2")
plt.title('クラス0〜2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_3=pd.read_csv('./data/line_data.csv')
plt.grid()
plt.plot('期間','件数',data=df_3,c='royalblue',marker='o')
plt.tick_params(axis='x',labelrotation=45)
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
```

```python
plt.grid()
plt.bar('期間','費用',data=df_3,color='royalblue',align='center')
plt.tick_params(axis='x',labelrotation=45)
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
```
